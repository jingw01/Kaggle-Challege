You must predict the total count of bikes rented during each hour covered by the test set, using only information available prior to the rental period.

#### Setup and Plotting code

```python
#Source --- CS109; 

%matplotlib inline
from collections import defaultdict
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from matplotlib import rcParams
import matplotlib.cm as cm
import matplotlib as mpl

#colorbrewer2 Dark2 qualitative color table
dark2_colors = [(0.10588235294117647, 0.6196078431372549, 0.4666666666666667),
                (0.8509803921568627, 0.37254901960784315, 0.00784313725490196),
                (0.4588235294117647, 0.4392156862745098, 0.7019607843137254),
                (0.9058823529411765, 0.1607843137254902, 0.5411764705882353),
                (0.4, 0.6509803921568628, 0.11764705882352941),
                (0.9019607843137255, 0.6705882352941176, 0.00784313725490196),
                (0.6509803921568628, 0.4627450980392157, 0.11372549019607843)]

rcParams['figure.figsize'] = (10, 6)
rcParams['figure.dpi'] = 150
rcParams['axes.color_cycle'] = dark2_colors
rcParams['lines.linewidth'] = 2
rcParams['axes.facecolor'] = 'white'
rcParams['font.size'] = 14
rcParams['patch.edgecolor'] = 'white'
rcParams['patch.facecolor'] = dark2_colors[0]
rcParams['font.family'] = 'StixGeneral'


def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecesasry plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
        
pd.set_option('display.width', 500)
pd.set_option('display.max_columns', 100)
```

```python
train=pd.read_csv("train.csv")

index=pd.DatetimeIndex(train['datetime'])
train.reset_index(index)
train['hour']=index.hour
train['date']=index.date
train['dayofweek']=index.dayofweek
varlist=list(train.columns.values)
train.head()
```

#### Explore the data

```python
#in group level, check whether there are correlation between some variables
from pandas.tools.plotting import scatter_matrix
fig=plt.figure()
time=range(24)
scatter_matrix(train[['weather','temp','atemp','humidity','windspeed','count']])
```

```python
#Explore relationship on everyday level
day=train.groupby('date')
daysum=day['casual','registered','count'].sum()
sum_list=list(daysum.columns.values)
renamedict_sum={k:k+"_daysum" for k in sum_list}
daysum.rename(inplace=True,columns=renamedict_sum)
daymean=day.mean()
daymean=daymean.drop(['casual','registered','count'],axis=1)
mean_list=list(daymean.columns.values)
#renamedict_mean={k:k+"_daymean" for k in mean_list}
#daymean.rename(inplace=True,columns=renamedict_sum)
daydf=daysum.join(daymean)
daydf.head()
```

```python
fig=plt.figure()
time=range(24)
fig,axes=plt.subplots(figsize=(10,10),nrows=4, ncols=3)
axes[0][0].plot(daydf['casual_daysum'],daydf['season'],'.')
axes[1][0].plot(daydf['casual_daysum'],daydf['holiday'],'.')
axes[2][0].plot(daydf['casual_daysum'],daydf['workingday'],'.')
axes[3][0].plot(daydf['casual_daysum'],daydf['weather'],'.')
axes[0][1].plot(daydf['season'],daydf['registered_daysum'],'.')
axes[1][1].plot(daydf['holiday'],daydf['registered_daysum'],'.')
axes[2][1].plot(daydf['workingday'],daydf['registered_daysum'],'.')
axes[3][1].plot(daydf['weather'],daydf['registered_daysum'],'.')
axes[0][2].plot(daydf['season'],daydf['count_daysum'],'.')
axes[1][2].plot(daydf['holiday'],daydf['count_daysum'],'.')
axes[2][2].plot(daydf['workingday'],daydf['count_daysum'],'.')
axes[3][2].plot(daydf['weather'],daydf['count_daysum'],'.')
```
