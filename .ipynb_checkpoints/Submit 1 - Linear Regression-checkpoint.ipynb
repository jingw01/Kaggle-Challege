# Bike Sharing Demand

Task: 

You are provided hourly rental data spanning two years. For this competition, the training set is comprised of the first 19 days of each month, while the test set is the 20th to the end of the month. You must predict the total count of bikes rented during each hour covered by the test set, using only information available prior to the rental period.

Data:http://www.kaggle.com/c/bike-sharing-demand/data

#### Setup and Plotting code

```python
#Source --- CS109; 

%matplotlib inline
from collections import defaultdict
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from matplotlib import rcParams
import matplotlib.cm as cm
import matplotlib as mpl

import brewer2mpl
from matplotlib import rcParams

#colorbrewer2 Dark2 qualitative color table
dark2_cmap = brewer2mpl.get_map('Dark2', 'Qualitative', 7)
dark2_colors = dark2_cmap.mpl_colors

rcParams['figure.figsize'] = (10, 6)
rcParams['figure.dpi'] = 150
rcParams['axes.color_cycle'] = dark2_colors
rcParams['lines.linewidth'] = 2
rcParams['axes.facecolor'] = 'white'
rcParams['font.size'] = 14
rcParams['patch.edgecolor'] = 'white'
rcParams['patch.facecolor'] = dark2_colors[0]
rcParams['font.family'] = 'StixGeneral'


def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecesasry plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
        
pd.set_option('display.width', 500)
pd.set_option('display.max_columns', 100)
```

#### Regression Model

```python
def predata(ds):
    indext=pd.DatetimeIndex(ds['datetime'].values)
    ds=ds.reset_index(indext)
    ds['hour']=indext.hour
    ds['date']=indext.date
    ds['month']=indext.month
    ds['year']=indext.year
    ds['dayofweek']=indext.dayofweek
    
    return ds
```

```python
train=pd.read_csv("train.csv")
train=predata(train)

test=pd.read_csv("test.csv")
test=predata(test)
```

```python
train=train.dropna()
small=train[['registered','hour','dayofweek','season','atemp','humidity','holiday','windspeed','temp']]
def createdummy(ds,outds):
    for i in range(4):
        outds['season'+str(i+1)]=ds['season']==i+1
        outds['year11']=ds['year']=2011
    for i in range(7):
        outds['day'+str(i+1)]=ds['dayofweek']==i+1
    outds['hour5-10']=[1 if i<10 & i>=5 else 0 for i in ds['hour'] ]
    outds['hour10-15']=[1 if i<15 & i>=10 else 0 for i in ds['hour'] ]
    outds['hour15-20']=[1 if i<20 & i>=15 else 0 for i in ds['hour'] ]
    outds['hour>20']=[1 if i>=20 else 0 for i in ds['hour'] ]
    return outds

small=createdummy(train,small)
small.head()
```

```python
Y=train[['casual']].values
X=small[['hour','day1','day2','day3','day5','day6','day7','season1','season2',
         'season4','hour5-10','hour10-15','hour15-20','hour>20','temp','humidity']].values
```

```python
from sklearn import linear_model
clf=linear_model.LinearRegression()
clf.fit(X,Y)
clf.score(X,Y)
```

```python
x_test=test[['hour','temp','humidity','holiday','atemp']]
x_test=createdummy(test,x_test)
xtest=x_test[['hour','day1','day2','day3','day5','day6','day7','season1','season2',
         'season4','hour5-10','hour10-15','hour15-20','hour>20','temp','humidity']].values
```

```python
predicted_testc=clf.predict(xtest)
```

```python
Y=train[['registered']].values
X=small[['hour','season1','season3','season4','hour5-10','hour10-15','hour15-20','hour>20','holiday','atemp','humidity']].values
clf=linear_model.LinearRegression()
clf.fit(X,Y)
clf.score(X,Y)
```

```python
xtest2=x_test[['hour','season1','season3','season4','hour5-10','hour10-15','hour15-20','hour>20','holiday','atemp','humidity']].values
predict_testr=clf.predict(xtest2)
```

```python
test['casual']=[round(i) if i>=0 else 0 for i in predicted_testc]
test['registered']=[round(i) if i>=0 else 0 for i in predict_testr]
test['count']=test['casual']+test['registered']

```

```python
submit=test[['datetime','count']]
submit.to_csv('submit.csv')
```
